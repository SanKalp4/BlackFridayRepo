```python
df
```

```python
X= df[['User_ID','Product_ID','Gender','Occupation', 'Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3','0-17', '18-25','26-35','36-45','46-50','51-55','A','B','C','0','1','2','3','4+']]
Y= df['Purchase']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test, Y_train, Y_test= train_test_split(X,Y, test_size = 0.3 , random_state= 1)
```

```python
print(X_train.size,X_test.size)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression().fit(X_train, Y_train)
```

```python
predictions= model1.predict(X_test)
np.set_printoptions(suppress= True)
print('Predicted_Labels:',(predictions)[:10])
print('Actual Labels:' ,  Y_test[:10])
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse =mean_squared_error(Y_test, predictions)
print('MSE:',mse)
rmse = np.sqrt(mse)
print('RMSE:', rmse)
r2 = r2_score(Y_test,predictions)
print('R2:', r2)
```
